# TASK 6. PERFORM TEXT DATA CLEANING BY REMOVING PUNCTUATIONS

```python
# String module is super useful when dealing with text data
# String contains constants and classes for working with text
import string
string.punctuation
```

```python
Test = '$I love Pandas &Data Analytics!!'
```

```python
Test_punc_removed = [char for char in Test if char not in string.punctuation]
Test_punc_removed
```

```python
# Join the characters again to form the string.
Test_punc_removed_join = ''.join(Test_punc_removed)
Test_punc_removed_join
```

```python
# Let's define a function to remove punctuations
def remove_punc(message):
    Test_punc_removed = [char for char in message if char not in string.punctuation]
    Test_punc_removed_join = ''.join(Test_punc_removed)
    
    return Test_punc_removed_join
```

```python
# Let's remove punctuations from our dataset 
df['verified_reviews_nopunc'] = df['verified_reviews'].apply(remove_punc)
```

```python
import nltk
from nltk.corpus import stopwords
import gensim
from gensim.utils import simple_preprocess
```

Right now I wanted to remove Stopwords.
And these are common words in the english
language that don't add a lot of value to us
especially when we train machine learning models.
And that's why in general we tend to remove them.

```python
# download stopwords
nltk.download('stopwords')
stopwords.words('english')
```

```python
# Add more stopwords by using extend
stop_words = stopwords.words('english')
stop_words.extend(['Amazon', 'amazon', 'alexa', 'echo', 'Alexa', 'device', 'Dot', 'dot'])
```

When you perform the simple pre process within the gensim library, all what it does,
is it takes any review like any text data and simply returns back
a list of cleaned up words.
Three terms like love my an echo for example, basically it
just takes all the keywords other words and then convert them
into lower case and then returned them back in the form
of a list.

```python
# Simple_preprocess converts a string into a series of lowered case tokens
# Let's try it on a sample dataset 
gensim.utils.simple_preprocess(df['verified_reviews'][0])
```

```python
# Remove stopwords and remove short words (less than 2 characters)
def preprocess(text):
    result = []
    for token in gensim.utils.simple_preprocess(text):
        if len(token) >= 1 and token not in stop_words:
            result.append(token)
    return result
```

Basically the function I'm going to call it pre process.
The function is going to take text data and all
but the function is going to do is going to create a for loop
that go through every single word that I have in my text data
and here I wanted to do two things I wanted to make sure
first of all that if the length of the token is greater
than one meaning that if I have any word that has More
than one character and the token is not in stop words meaning
if the token does not belong to any of these common words
and does not belong to any of the extended as well.
Common words, that's good.
Please go ahead, grab that and append the token to my result
and return back the result for me.

```python
# apply pre-processing to the text column
df['verified_reviews_nopunc_nostopwords'] = df['verified_reviews_nopunc'].apply(preprocess)
```
